```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

VA_pop_dat = pd.read_csv('VA_populations.csv')

counties_pop = dict(zip(VA_pop_dat['subregion'], VA_pop_dat['pop']))

words = WordCloud(background_color = 'white').generate_from_frequencies(VA_pop_dat['pop'])
plt.figure(figsize = (12,6))
plt.imshow(words, interpolation = 'auto')
```
